# 1. Import Libraries and Load Data

```python
import torch
import torchvision
import torchvision.transforms as transforms
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os

# Load datasets
train_dir = 'output_dataset/train'
val_dir = 'output_dataset/val'
test_dir = 'output_dataset/test'

train_dataset = ImageFolder(root=train_dir)
val_dataset = ImageFolder(root=val_dir)
test_dataset = ImageFolder(root=test_dir)
```

# 2. Class Distribution Analysis

```python
#Analyze the distribution of classes in the training, validation, and test datasets to ensure a balanced dataset.
def plot_class_distribution(dataset, title):
    class_counts = {}
    for _, label in dataset:
        class_name = dataset.classes[label]
        class_counts[class_name] = class_counts.get(class_name, 0) + 1

    plt.bar(class_counts.keys(), class_counts.values())
    plt.title(title)
    plt.xlabel('Class')
    plt.ylabel('Number of Images')
    plt.show()

plot_class_distribution(train_dataset, 'Training Dataset Class Distribution')
plot_class_distribution(val_dataset, 'Validation Dataset Class Distribution')
plot_class_distribution(test_dataset, 'Test Dataset Class Distribution')
```

# 3. Sample Image Visualization

```python
#Visualize sample images from each class to get a sense of the data quality and variability.
def show_sample_images(dataset, num_samples=5):
    class_names = dataset.classes
    for class_name in class_names:
        class_idx = dataset.class_to_idx[class_name]
        class_images = [img for img, label in dataset if label == class_idx]
        plt.figure(figsize=(10, 5))
        plt.title(class_name)
        for i in range(num_samples):
            plt.subplot(1, num_samples, i+1)
            plt.imshow(class_images[i])
            plt.axis('off')
        plt.show()

show_sample_images(train_dataset)
```

```python
import matplotlib.pyplot as plt
from torchvision.datasets import ImageFolder

# Load the training dataset
train_dir = 'output_dataset/train'
train_dataset = ImageFolder(root=train_dir)

# Function to display one image from each class
def show_one_image_per_class(dataset):
    class_names = dataset.classes
    plt.figure(figsize=(10, 5))
    for i, class_name in enumerate(class_names):
        # Find the first image belonging to the current class
        class_idx = dataset.class_to_idx[class_name]
        for img, label in dataset:
            if label == class_idx:
                plt.subplot(1, len(class_names), i + 1)
                plt.imshow(img)
                plt.title(class_name)
                plt.axis('off')
                break  # Break after displaying one image from the class
    plt.show()

# Display one image from each class
show_one_image_per_class(train_dataset)
```
